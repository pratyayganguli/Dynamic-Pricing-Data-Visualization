#### Monthly Data Function Prototypes ###

```python
def plotDFLimit(dataFrame, kind, x, y, rowLimit):
    dataFrame[:rowLimit].plot(kind = kind, xlabel = x, ylabel = y )

def plotDF(dataFrame, kind, x, y):
    dataFrame.plot(kind = kind, xlabel = x, ylabel = y )

def plotDFFrom(dataFrame, kind, x, y, _from):
    dataFrame[_from:].plot(kind = kind, xlabel = x, ylabel = y )

def plotDFRange(dataFrame, kind, x, y, _from, _to):
    dataFrame[_from:_to+1].plot(kind = kind, xlabel = x, ylabel = y )
```

```python
PATH = '../datasets/csv/'

FILE_NAME = {
    'monthly': 'monthly-booking-data',
    'daily': 'daily-booking-data',
    'annually' : 'annual-booking-data'
}

FILE_EXTENSION = '.csv'

csvFile = open(PATH + FILE_NAME['monthly'] + FILE_EXTENSION)

dailyDF = pd.read_csv(csvFile, names = [
    '08:00 - 10:00',
    '10:00 - 12:00',
    '12:00 - 14:00',
    '14:00 - 16:00',
    '16:00 - 18:00',
    '18:00 - 20:00'
])

plotDFLimit(dailyDF, 'line', 'Time Block', 'Number of Bookings', 3) 
plotDFRange(dailyDF, 'bar', 'Time Block', 'Number of Bookings', 1, 6)
```
