### Importing the Libraries ###

```python
import numpy as np
import random
import time
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import clear_output
```

### Generating Annual Dataset ###


```python
annualData = np.array([[0, 0, 0, 0, 0, 0]])
month = 0
MAX_MONTH = 12
    
csvFile = open('../datasets/csv/annual-booking-data.csv', 'w')

while month < 12:
    writer = csv.writer(csvFile)
    annualData = np.append(annualData, np.array([[
                    random.randint(700, 1000),
                    random.randint(600, 900),
                    random.randint(400, 700),
                    random.randint(500, 700),
                    random.randint(700, 1100),
                    random.randint(1000, 1400)
                ]]), axis = 0)
    month+=1
    writer.writerow(annualData[month])

csvFile.close()
print('File Exported')

annualDataFrame = pd.DataFrame(annualData, columns = [
    '08:00-10:00',
    '10:00-12:00',
    '12:00-14:00',
    '14:00-16:00',
    '16:00-18:00',
    '18:00-20:00'
])
annualDataFrame = annualDataFrame[1:]
annualDataFrame
```

### Generating Monthly Data ###

```python
monthlyData = np.array([[0, 0, 0, 0, 0, 0]])
days = 0
MAX_DAYS = 28
    
csvFile = open('../datasets/csv/monthly-booking-data.csv', 'w')

while days < 28:
    writer = csv.writer(csvFile)
    monthlyData = np.append(monthlyData, np.array([[
                    random.randint(700, 1000),
                    random.randint(600, 900),
                    random.randint(400, 700),
                    random.randint(500, 700),
                    random.randint(700, 1100),
                    random.randint(1000, 1400)
                ]]), axis = 0)
    days+=1
    writer.writerow(monthlyData[days])

csvFile.close()
print('File Exported')

monthlyDataFrame = pd.DataFrame(monthlyData, columns = [
    '08:00-10:00',
    '10:00-12:00',
    '12:00-14:00',
    '14:00-16:00',
    '16:00-18:00',
    '18:00-20:00'
])

monthlyDataFrame = monthlyDataFrame[1:]
monthlyDataFrame
```

```python
annualDataFrame.plot(kind = 'line', xlabel = 'Days', ylabel = 'Bookings')
monthlyDataFrame.plot(kind = 'kde', xlabel = 'Days', ylabel = 'Bookings')
```
