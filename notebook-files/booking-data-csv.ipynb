### Importing the Libraries ###

```python
import numpy as np
import random
import time
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import clear_output
```

### Generating Monthly Dataset ###


```python
monthlyData = np.array([[0, 0, 0, 0, 0, 0]])
month = 0
MAX_MONTH = 12
    
csvFile = open('../datasets/csv/monthly-booking-data.csv', 'w')

while month < 12:
    writer = csv.writer(csvFile)
    monthlyData = np.append(monthlyData, np.array([[
                    random.randint(100, 200),
                    random.randint(300, 500),
                    random.randint(400, 600),
                    random.randint(100, 200),
                    random.randint(300, 500),
                    random.randint(400, 600)
                ]]), axis = 0)
    month+=1
    writer.writerow(monthlyData[month])

csvFile.close()
print('File Exported')

monthlyDataFrame = pd.DataFrame(monthlyData, columns = [
    '08:00-10:00',
    '10:00-12:00',
    '12:00-14:00',
    '14:00-16:00',
    '16:00-18:00',
    '18:00-20:00'
])
monthlyDataFrame = monthlyDataFrame[1:]
monthlyDataFrame
```
