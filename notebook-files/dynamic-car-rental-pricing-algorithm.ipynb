### Read operations - Annual Booking Dataset ###

#### Plot Functions ###

```python
def plotLimit(dataFrame, kind, x, y, rowLimit):
    dataFrame[:rowLimit].plot(kind = kind, xlabel = x, ylabel = y )

def plotDF(dataFrame, kind, x, y):
    dataFrame.plot(kind = kind, xlabel = x, ylabel = y )

def plotFrom(dataFrame, kind, x, y, _from):
    dataFrame[_from:].plot(kind = kind, xlabel = x, ylabel = y )

def plotRange(dataFrame, kind, x, y, _from, _to):
    dataFrame[_from:_to+1].plot(kind = kind, xlabel = x, ylabel = y )

def plotByYear(dataFrame, kind, x, y, monthIndex):
    dataFrame.iloc[monthIndex].plot(kind = kind, xlabel = x, ylabel = y)
    
def plotDensityByYear(dataFrame, monthIndex):
    dataFrame.iloc[monthIndex].plot(kind = 'kde')
```

#### Data Operation Functions ####

```python
def getDataByBlock(dataFrame, timeBlock):
    return np.array([dataFrame[timeBlock]])

def grossBookingsByBlock(dataFrame, timeBlock):
    return np.sum(np.array([dataFrame[timeBlock]]))

def grossBookingsByMonth(dataFrame, monthIndex):
    return np.sum(np.array([dataFrame.iloc[monthIndex]]))

def minBookingByMonth(dataFrame, monthIndex):
    return np.amin(np.array([dataFrame.iloc[monthIndex]]))

def maxBookingByMonth(dataFrame, monthIndex):
    return np.amax(np.array([dataFrame.iloc[monthIndex]])) 
```

#### Predictive Analysis Functions ####
