```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib notebook
```

```python
file = open('../datasets/csv/distance-data.csv', 'r')
data = pd.read_csv(file)
file.close()
data = data.drop(data.columns[0], axis = 1)
data.head()
```

```python
data.plot(kind = 'scatter', x = 'Age', y = 'Distance', c = 'g')
plt.show()
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(data.Age, data.Distance)
plt.scatter(xTrain, yTrain, label = 'Distance Travelled Training Data', color = 'b', alpha = .7)
plt.scatter(xTest, yTest, label = 'Distance Travelled Test Data', color = 'red', alpha = .7)
plt.legend()
plt.show()
```

```python
age = data['Age'].values
distance = data['Distance'].values

meanAge = np.mean(age)
meanDistance = np.mean(distance)

n = 0
d = 0

for i in range(0, len(age)):
    n += (age[i] - meanAge) * (distance[i] - meanDistance)
    d += (age[i] - meanAge) ** 2

slope = n/d
constant = meanDistance - (slope*meanAge)
slope, constant
```

```python
ageMax = np.max(age)
ageMin = np.min(age) 
x = np.linspace(ageMin, ageMax)
y = slope * x + constant
```

```python
plt.plot(x, y , color = 'green', label = 'regression line')
plt.scatter(age, distance, c = 'red', label= 'scatter data')
plt.xlabel('Age')
plt.ylabel('Distance')
plt.legend
plt.show()
```

```python
totalSumSq = 0
totalSumSqResidual = 0

for i in range(0, len(age)):
    yPred = slope * age[i] + constant
    totalSumSq += (distance[i] - meanDistance) ** 2
    totalSumSqResidual += (distance[i] - yPred) ** 2

R = 1 - (totalSumSqResidual/totalSumSq)
R
```
