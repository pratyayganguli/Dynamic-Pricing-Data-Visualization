```python
import random
import numpy as np
import pandas as pd
```

```python
timeHrs = np.array([])
timeMins = np.array([])
carModels = np.array(['Swift', 'Wagonr', 'Dzire', 'i10', 'XUV 500', 'Celerio', 'Baleno', 'Tiago', 'Verna'])
distanceInKms = np.array([])

MAX_ROWS = 10

row = 0

while row < MAX_ROWS:
    timeHrs = np.append(timeHrs, int(random.randint(8, 20)))
    row += 1

row = 0

while row < MAX_ROWS:
    timeMins = np.append(timeMins, int(random.randint(10, 59)))
    row += 1
    
row = 0

while row < MAX_ROWS:
    distanceInKms = np.append(distanceInKms, int(random.randint(180, 210)))
    row += 1
```

```python
npData = np.array([
    ['', '', '', '']
])

npData = np.append(npData, np.array([
    [
        timeHrs[random.randint(0, len(timeHrs)-1)],
        timeMins[random.randint(0, len(timeMins)-1)],
        carModels[random.randint(0, len(carModels)-1)],
        distanceInKms[random.randint(0, len(distanceInKms)-1)]

    ]
]), axis=0)

npData = npData[1:]
```

```python
MAX_ROWS = 500
row = 0

while row < MAX_ROWS:
    npData = np.append(npData, np.array([
        [
            timeHrs[random.randint(0, len(timeHrs)-1)],
            timeMins[random.randint(0, len(timeMins)-1)],
            carModels[random.randint(0, len(carModels)-1)],
            distanceInKms[random.randint(0, len(distanceInKms)-1)]

        ]
    ]), axis=0)
    row += 1

dataFrame = pd.DataFrame(data = npData, columns = [
    'HH', 'MM', 'Car Model', 'Distance'
])

dataFrame
```

```python
file = open('../datasets/csv/data-monthly.csv', 'w')
dataFrame.to_csv(file)
file.close()
```
